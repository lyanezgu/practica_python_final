```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
```

# Ejercicio 1

## Cargo dataset 

```python
cwd = Path.cwd()
#crea en path relativo de donde esta el csv
filename = os.path.join(cwd, "Fitness_trackers.csv") 
data_df = pd.read_csv(filename)

data_df.head(n=20)
```

```python
print('Cantidad de Filas y columnas:',data_df.shape)
print('Nombre columnas:',data_df.columns)
```

## Estadísticos del dataset 

```python
#Número de muestras (valores distintos de missing)
data_df.info()
```

```python
#cambio de tipo string a float estas columnas

data_df["Selling Price"]=data_df["Selling Price"].str.replace(',','').astype(float)

data_df["Original Price"]=data_df["Original Price"].str.replace(',','').astype(float)

data_df["Reviews"]=data_df["Reviews"].str.replace(',','').astype(float)
```

```python
#número de muestras y NAs
data_df.info()
```

```python
#calculamos los principales estadísticos de las variables numéricas
data_df.describe()
```

# Ejercicio2

## Ejercicio 2.1 (por dispositivo)

```python
data_df.groupby('Device Type').agg({'Selling Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Original Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Rating (Out of 5)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Average Battery Life (in days)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                   'Reviews':['count',max,min,'var','median',lambda x: x.isnull().sum()]})

```

## Ejercicio 2.2 ( por precio de venta)

```python
venta = data_df.groupby('Device Type')['Selling Price'].mean()

print(venta)

df_fb = data_df.loc[data_df['Device Type'] == 'FitnessBand']

#dataframe de fitness band menores e iguales a la media selling price
df_fb1=df_fb.loc[df_fb['Selling Price'] <= venta['FitnessBand']]

df_fb1.groupby('Device Type').agg({'Selling Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Original Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Rating (Out of 5)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Average Battery Life (in days)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                   'Reviews':['count',max,min,'var','median',lambda x: x.isnull().sum()]})
```

```python
#dataframe de fitness band mayores a la media selling price
df_fb2 = df_fb.loc[df_fb['Selling Price'] > venta['FitnessBand']]

df_fb2.groupby('Device Type').agg({'Selling Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Original Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Rating (Out of 5)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Average Battery Life (in days)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                   'Reviews':['count',max,min,'var','median',lambda x: x.isnull().sum()]})
```

```python

#dataframe de smartwatch band menores e iguales a la media selling price

df_sw = data_df.loc[data_df['Device Type'] == 'Smartwatch']

df_sw1=df_sw.loc[df_sw['Selling Price'] <= venta['Smartwatch']]

df_sw1.groupby('Device Type').agg({'Selling Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Original Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Rating (Out of 5)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Average Battery Life (in days)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                   'Reviews':['count',max,min,'var','median',lambda x: x.isnull().sum()]})
```

```python
#dataframe de smartwatch band mayores a la media selling price

df_sw2= df_sw.loc[df_sw['Selling Price'] > venta['Smartwatch']]

df_sw2.groupby('Device Type').agg({'Selling Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Original Price':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Rating (Out of 5)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                    'Average Battery Life (in days)':['count',max,min,'var','median',lambda x: x.isnull().sum()],
                                   'Reviews':['count',max,min,'var','median',lambda x: x.isnull().sum()]})
```
