# EJERCICIO 3

 Añadimos una nueva columna (12 columnas tendremos ahora en el conjunto de datos) que contenga el ratio de duración/precio indicado que nos interesa

```python
data_df_ratio = data_df.assign(ratio_duracion = data_df['Average Battery Life (in days)']*1000/data_df['Selling Price'])
print(data_df_ratio)
```

Ahora pasamos al cálculo del valor de la media para ese ratio de duración/precio obtenido

```python
ratio_mean = data_df_ratio['ratio_duracion'].mean()
print(ratio_mean)
```



Sacamos los dispositivos en los que la ratio de la duración de la batería con respecto al precio sea superior a la media de ratio de duración indicada (0.99)



```python
df_ratioSup = data_df_ratio.loc[data_df_ratio['ratio_duracion'] > ratio_mean]
print(df_ratioSup)
```

Como conclusión podemos decir que la media del ratio de duración/precio es 0.99 y de los 565 dispositivos que hay en total, 158 tienen el ratio de duración superior a esa media. Esto supone que apenas el 28% de los dispositivos, contando con fitnessband y smartwatch, superan la media del ratio duración/precio. De estos 28%, aproximadamente un 70% son Smartwatches, 19% son da marca huami y 17% da marca FOSSIL (dos mayores frecuencias de marca) y casi un 45% tienen el cuerpo de la correa “Black”.

## 7.1. Estudio y representación de la correlación

Estudiamos la correlación entre las variables y la representamos a través de un mapa de calor que nos facilite su visualización

```python
corr_matrix = data_df_ratio.corr(method='pearson')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))

sns.heatmap(
    corr_matrix,
    annot     = True,
    cbar      = False,
    annot_kws = {"size": 8},
    vmin      = -1,
    vmax      = 1,
    center    = 0,
    cmap      = sns.diverging_palette(20, 220, n=200),
    square    = True,
    ax        = ax
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right',
)

ax.tick_params(labelsize = 10)
```

## 7.2. ¿Existe alguna correlación que llame especialmente la atención?

- Destaca que las variables económicas de precios (Selling y Original Price) estan muy poco correlacionadas (0.14 y 0.19 respectivamente), es decir, mayores precios no implican dispositivos con mayor duración de batería entre carga y carga. Esto en el fondo se da por que al ser dispositivos de poco tamaño, las baterias que se le pueden poner no son muy grandes, y, por ello, los dispositivos caros al tener mayores funcionalidades, también consumen más batería y su duración final se ve mermada.

- También destaca, con una correlación muy baja (0.1), que a mayor número de Reviews realizadas no se incrementa el Rating (Out of 5) o puntuación de los consumidores dado a ese dispositivo.

- Con el ratio duración/precio calculado, podemos ver que a mayores valores del mismo, esta correlacionado con el precio de venta de forma negativa (a niveles medios de correlación con un valor negativo de -0.45). Esto implica que en general valores mayores de este ratio llevan a precios de dispositivos más bajos. Esto en el fondo tiene sentido, por que las FitnessBand (dispositivos vistos anteriormente como los que mejor ratio de duración/precio tienen en media) son respecto a los smartwatch más baratos, con menos funciuonalidades, pero con mucha mejor duración de bateria.

- Por último también destacar la escasa correlación que muestra la variable Average Battery Life (in days) con respecto al Rating (Out of 5) o valoración de los compradores (correlación de -0.16). Esto se podría entender como que mejor duración de batería no implica siempre mejores valoraciones del producto.

## 9.4. Análisis de regresión lineal

Analizamos a través de una regresión lineal la variable precio de venta en función de la duración de la bateria de los dispositivos

```python
plt.figure(figsize = (15,10))
 
# plot
sns.regplot(x=data_df["Average Battery Life (in days)"], y=data_df["Selling Price"], line_kws={"color":"r","alpha":0.7,"lw":5})
plt.show()
```

Como se puede ver en el gráfico de puntos no existe una relación lineal clara entre las variables de precio de venta y la media de vida de la batería. Se observan casos con media de vida de la batería de 1 día con precios elevados, pero también casos media de vida de la batería de más de 40 días y con precios elevados. La regresión lineal mostrada en rojo no muestra un patron de relación concluyente entre ambas variables, es decir, a mayor precio la media de vida de la batería no siempre es alta.
