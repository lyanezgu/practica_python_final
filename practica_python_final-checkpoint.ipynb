# EJERCICIO 3

```python
# Añadimos una nueva columna que contenga el ratio de duración indicado que nos interesa
data_df_ratio = data_df.assign(ratio_duracion = data_df['Average Battery Life (in days)']*1000/data_df['Selling Price'])
print(data_df_ratio)
```

```python
# Calculamos el valor de la media para ese ratio de duración
ratio_mean = data_df_ratio['ratio_duracion'].mean()
print(ratio_mean)
```

```python
# Sacamos los dispositivos en los que la ratio de la duración de la batería con respecto al precio sea superior a la media de ratio de duración indicada
df_ratioSup = data_df_ratio.loc[data_df_ratio['ratio_duracion'] > ratio_mean]
print(df_ratioSup)
```

```python
dispositivo_ratio = data_df_ratio_orden.groupby('Device Type')['ratio_duracion'].mean()
print(dispositivo_ratio)
```

El mejor dispositivo según relación duración/precio es FitnessBand de Xiaomi Band - HRX Edition con un ratio del 15.396459.

También vemos que el ratio duración / precio en media es superior para los dispositivos de tipo FitnessBand frente a los Smartwatch.

```python
marca_ratio = data_df_ratio_orden.groupby('Brand Name')['ratio_duracion'].mean()
print(marca_ratio)
```

Según el analsis del ratio duración/precio medio por marca, Xiaomi destacaría por encima del resto de marcas con una media de ratio del 5.58. También son destacables marcas como LAVA, Honor, Fastrack.

regresión lineal 

```python
plt.figure(figsize = (15,10))
 
# plot
sns.regplot(x=data_df["Average Battery Life (in days)"], y=data_df["Selling Price"], line_kws={"color":"r","alpha":0.7,"lw":5})
plt.show()
```

Como se puede ver en el gráfico de puntos no existe una relación lineal clara entre las variables de precio de venta y la media de vida de la batería. Se observan casos con media de vida de la batería de 1 día con precios elevados, pero también casos media de vida de la batería de más de 40 días y con precios elevados. La regresión lineal mostrada en rojo no muestra un patron de relación concluyente entre ambas variables, es decir, a mayor precio la media de vida de la batería no siempre es alta.
